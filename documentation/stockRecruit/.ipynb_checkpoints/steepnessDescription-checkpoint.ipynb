# Description of steepness formulations



### Data
1. $R_{y}$ Recruitment in year $y$
2. $SSB_{y}$ Spawner biomass in year $y$
3. $SSBR_{F=0}$ Spawner biomass-per-recruit at $F=0$ (calculated assuming selectivity, weight- and maturity-at-age and natural mortality parameters in the stock assessment model from which $R$ and $SSB$ were estimated)
4. $T^{*}_{y}$ Temperature anomaly in year $y$

### Estimated parameters
1. $h$ steepness
2. $R_{0}$ unfished recruitment
3. $\sigma^{2}$ variance of normally distributed errors
4. $\beta_{1}$ Temperature anomaly $h$ effect
5. $\beta_{2}$ Temperature anomaly $R_{0}$ effect
6. $\beta_{3}$ Temperature anomaly $\varepsilon$ effect


### Base model

$$
\begin{equation*}
\label{eq:eqn1}
R_{y} = \frac{4h\frac{SSB_{y-1}}{SSBR_{F=0}}}{(1+h)+(5h-1)\frac{SSB_{y-1}}{R_{0}SSBR_{F=0}}}e^{\varepsilon}
\end{equation*}
$$

where $\varepsilon \sim N(0,\sigma^{2})$


### Temperature anomaly impacts steepness
Replace $h$ with $he^{\beta_{1}T^{*}_{y}}$

$$
\begin{equation*}
\label{eq:eqn2}
R_{y} = \frac{4(he^{\beta_{1}T^{*}_{y}})\frac{SSB_{y-1}}{SSBR_{F=0}}}{(1+(he^{\beta_{1}T^{*}_{y}}))+(5(he^{\beta_{1}T^{*}_{y}})-1)\frac{SSB_{y-1}}{R_{0}SSBR_{F=0}}}e^{\varepsilon}
\end{equation*}
$$


### Temperature anomaly impacts unfished recruitment
Replace $R_{0}$ with $R_{0}e^{\beta_{2}T^{*}_{y}}$

$$
\begin{equation*}
\label{eq:eqn3}
R_{y} = \frac{4h\frac{SSB_{y-1}}{e^{\beta_{2}T^{*}_{y}}SSBR_{F=0}}}{(1+h)+(5h-1)\frac{SSB_{y-1}}{R_{0}e^{\beta_{2}T^{*}_{y}}SSBR_{F=0}}}e^{\varepsilon}
\end{equation*}
$$


### Temperature anomaly impacts errors
Replace $\varepsilon$ with $\varepsilon + \beta_{3}T^{*}_{y}$

$$
\begin{equation*}
\label{eq:eqn4}
R_{y} = \frac{4h\frac{SSB_{y-1}}{SSBR_{F=0}}}{(1+h)+(5h-1)\frac{SSB_{y-1}}{R_{0}SSBR_{F=0}}}e^{(\varepsilon + \beta_{3}T^{*}_{y})}
\end{equation*}
$$
